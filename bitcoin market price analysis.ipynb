```python
#splitting data 
from sklearn.model_selection import train_test_split
train, test = train_test_split(bitcoin_new, train_size= 0.7, test_size= 0.3,random_state= 1234)
```

```python
train.shape, test.shape
```

```python
#resceaaling the feature 
#max_min scalling 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
vars = ['btc_market_price','btc_total_bitcoins','btc_market_cap','btc_trade_volume','btc_blocks_size','btc_hash_rate',
                                     'btc_difficulty','btc_transaction_fees']
```

```python
train[vars] = scaler.fit_transform(train[vars])
```

```python
train.head()
```

```python
train.describe()
```

```python
plt.figure(figsize=(25,20))
sns.heatmap(train.corr(),annot = True)
```

```python
y_train  = train.pop('btc_market_price')
X_train = train
```

```python
y_train.shape, X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python

lm = LinearRegression()
```

```python
model = lm.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.score(X_train,y_train)  #### 100 % Accuracy
```

```python
#### Feature Selection Method which is called RFE-  Recurrsive Feature Elimination
from sklearn.feature_selection import RFE
```

```python
rfe = RFE(lm,12)
```

```python
rfe = rfe.fit(X_train,y_train)
```

```python
rfe.ranking_
```

```python
list(zip(X_train.columns,rfe.ranking_))
```

```python
y_test  = test.pop('btc_market_price')
X_test = test
```

```python
pred = model.predict(X_test)
```
